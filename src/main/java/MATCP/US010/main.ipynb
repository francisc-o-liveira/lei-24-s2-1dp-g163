```python
import pandas as pd
import matplotlib.pyplot as plt

# Read data from the CSV file (assuming the file is named "EquipmentUsed.csv")
file_path = "EquipmentUsed.csv"
data = pd.read_csv(file_path)

# Assuming the CSV file has a column named "Equipment" containing equipment names
# Count the occurrences of each equipment type
equipment_counts = data["Equipment"].value_counts()

# Calculate the total count of equipment usage
total_usage = equipment_counts.sum()

# Calculate the percentage of usage for each equipment type
equipment_percentages = (equipment_counts / total_usage) * 100

# Create a pie chart
plt.figure(figsize=(8, 6))
plt.pie(equipment_percentages, labels=equipment_percentages.index, autopct="%1.1f%%", startangle=90)
plt.title("Equipment Usage Percentage")
plt.axis("equal")  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```

# Explicações Teóricas
### Frquência Absoluta:
A frequência absoluta representa o número de vezes que um determinado valor ocorre em um conjunto de dados. Por outras palavras palavras, é a contagem direta de quantas vezes um valor específico aparece na amostra. No contexto presente a frequência absoluta de cada equipamento representa o número de utilizadores que utilzaram esse mesmo equipamento  num determinado dia.

### Frequência Relativa:
A frequência relativa representa a percentagem que um determinado evento ocorre em relação ao total de observações ou eventos. No contexto presente, a frequência relativa de cada equipamento representa a porcentagem de utilizadores que utilizaram esse equipamento em relação ao total de utilizadores. 

### Gráfico Circular:
Um gráfico circular é uma representação dos dados num círculo dividido em setores circulares cuja área (e ângulo ao centro correspondente) é proporcional à frequência da categoria/classe que representam.
