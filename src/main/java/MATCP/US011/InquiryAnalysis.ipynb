```python
import fileinput # Módulo para ler arquivos de texto
import matplotlib.pyplot as plt # Biblioteca para visualização de dados

# Contadores que armazenam o nº de respostas "Sim" ou "Não" (se recomendam ou não o parque a outras pessoas) para cada tipo de indivíduo 
Type1Yes = 0
Type2Yes = 0
Type3Yes = 0

Type1No = 0
Type2No = 0
Type3No = 0

# Listas que armazenam os tempos de uso para cada tipo de indivíduo
Type1TimesList = []
Type2TimesList = []
Type3TimesList = []

#As funções switch_caseYes e switch_caseNo incrementam os contadores e adicionam o tempo à lista correspondente com base na resposta "Sim" ou "Não" para cada tipo de indivíduo.
# Função para lidar com respostas "Sim".
def switch_caseYes(argument, argument2):
    #argument: o tipo de indivíduo ("Child", "Adult" ou "Senior"); 
    #argument2: o tempo de uso do parque pelo indivíduo.
    global Type1Yes, Type2Yes, Type3Yes, Type1TimesList, Type2TimesList, Type3TimesList
    if argument == "1":
        Type1Yes += 1
        Type1TimesList.append(argument2)
    elif argument == "2":
        Type2Yes += 1
        Type2TimesList.append(argument2)
    elif argument == "3":
        Type3Yes += 1
        Type3TimesList.append(argument2)
    else:
        print("Invalid case")
# Função para lidar com respostas "Não".
def switch_caseNo(argument, argument2):
    global Type1No, Type2No, Type3No, Type1TimesList, Type2TimesList, Type3TimesList
    if argument == "1":
        Type1No += 1
        Type1TimesList.append(argument2)
    elif argument == "2":
        Type2No += 1
        Type2TimesList.append(argument2)
    elif argument == "3":
        Type3No += 1
        Type3TimesList.append(argument2)
    else:
        print("Invalid case")

#Esta função exibe o menu principal do programa, listando as opções disponíveis para o usuário escolher.
def main_menu():
    print("1. Enter the data file name")
    print("2. Show proportions of recommendation by age group")
    print("3. Generate boxplots for monthly frequency of use")
    print("4. Exit")

#Esta função lê os dados de um arquivo CSV fornecido pelo usuário. Os dados são organizados por tipo de indivíduo ("Child", "Adult", "Senior") e
#por tipo de resposta ("Y" ou "N"). De seguida, são processados pelas funções switch_caseYes() e switch_caseNo().
def read_data():
    global Type1Yes, Type2Yes, Type3Yes, Type1No, Type2No, Type3No, Type1TimesList, Type2TimesList, Type3TimesList
    filename = input("Enter the data file name (.csv): ")
    scan = open(filename)
    lineNumber=-1
    for line in scan:
        lineNumber+=1
        if(lineNumber>0):
            arrayLine = line.split(";")
            if arrayLine[1] == 'Y':
                switch_caseYes(arrayLine[0], int(arrayLine[2]))
            else:
                switch_caseNo(arrayLine[0], int(arrayLine[2]))

#Esta função calcula e exibe o nº total de indivíduos por faixa etária, somando as aprovações e rejeições de cada grupo.
#Ou seja, determina o número total de indivíduos de cada tipo.
def show_proportions():
    Type1 = Type1Yes + Type1No
    Type2 = Type2Yes + Type2No
    Type3 = Type3Yes + Type3No

    #Proporçao de Individuos que recomendam o parque por cada grupo etario (desde que haja pelo menos um indivíduo desse tipo)
    if Type1 != 0:
        pType1Yes = (Type1Yes/Type1)
        print(f"The proportion of Children who would recommend the park to others: {pType1Yes}\n")

    if Type2 != 0:
        pType2Yes = (Type2Yes/Type2)
        print(f"The proportion of Adults who would recommend the park to others: {pType2Yes}\n")

    if Type3 != 0:
        pType3Yes = (Type3Yes/Type3)
        print(f"The proportion of Seniors who would recommend the park to others: {pType3Yes}\n")


#verifica se há dados disponíveis para criar um boxplot dos tempos de uso mensal para o grupo de indivíduos do tipo "Child", "Adult" e "Senior".
#Se houver dados disponíveis para um determinado grupo, um boxplot é criado usando a biblioteca matplotlib.
def generate_boxplots():
    if Type1Yes + Type1No != 0:
        plt.boxplot(Type1TimesList, patch_artist=True, boxprops=dict(facecolor='lightblue'))
        plt.title('Boxplot of monthly frequency of use of Children')
        plt.ylabel('Times')
        plt.show()

    if Type2Yes + Type2No != 0:
        plt.boxplot(Type2TimesList, patch_artist=True, boxprops=dict(facecolor='lightblue'))
        plt.title('Boxplot of monthly frequency of use of Adults')
        plt.ylabel('Times')
        plt.show()

    if Type3Yes + Type3No != 0:
        plt.boxplot(Type3TimesList, patch_artist=True, boxprops=dict(facecolor='lightblue'))
        plt.title('Boxplot of monthly frequency of use of Seniors')
        plt.ylabel('Times')
        plt.show()


#O menu principal é apresentado, permitindo que o usuário faça uma escolha e seja executada a ação correspondente.
while True:
    main_menu()
    choice = input("Enter your choice: ")
    #Verifica se o usuário escolheu a opção 1, que é para ler os dados do arquivo.
    if choice == "1":
        read_data()
    #Verifica se o usuário escolheu a opção 2, que é para mostrar as proporções de recomendação por grupo etário.    
    elif choice == "2":
        show_proportions()
    #Verifica se o usuário escolheu a opção 3, que é para gerar os boxplots dos tempos de uso mensal.
    elif choice == "3":
        generate_boxplots()
    #Verifica se o usuário escolheu a opção 4, que é para sair do programa. 
    #Se escolhida, uma mensagem de saída é exibida e o loop é encerrado com break.
    elif choice == "4":
        print("Exiting...")
        break
    #caso nenhuma das opções pedidas seja introduzida, é apresentada mensagem de erro.    
    else:
        print("Invalid choice. Please enter a valid option.")
```

# Explicação teórica: Boxplot

O boxplot é uma ferramenta gráfica que analisa as características de um conjunto de dados, incluindo a localização, assimetria, dispersão, comprimento da cauda e presença de outliers (medidas discrepantes).

No boxplot, uma caixa é desenhada do primeiro ao terceiro quartil. A linha horizontal dentro da caixa, corresponde à mediana. Fora da caixa, temos o mínimo,o máximo e possíveis outliers. O eixo x representa os dados a serem plotados, enquanto o eixo y mostra a distribuição de frequência. Desse modo, representa graficamente os extremos e os quartis de um determinado conjunto de dados
O primeiro quartil (Q1) indica o valor abaixo do qual estão os 25% inferiores dos dados, pelo que o terceiro quartil (Q3) indica o valor abaixo do qual estão os 75% inferiores dos dados. A mediana (ou segundo quartil, Q2) é representada por uma linha que passa pelo centro da caixa e divide em duas parcelas. 

A dispersão dos dados é representada pela amplitude do gráfico, que é calculada através da diferença entre o 3º e o 1º quartil. Quanto maior a amplitude, maior a variação nos dados.
A simetria é observada através da mediana. Caso a mediana se encontre no centro do retângulo, é simétrica. Caso contrário é assimétrica.
Quando a linha da mediana está próxima do primeiro quartil, os dados são assimétricos positivos. 
Se estiver próxima do terceiro quartil, os valores são assimétricos negativos.
As linhas que vão do retângulo até ao valor máximo e mínimo indica os comprimentos das caudas da distribuição.
Os outliers são os valores discrepantes que se encontram abaixo ou acima dos valores mínimos e máximos, respetivamente.

O limite de detecção de outliers é dado pela distância entre o primeiro e o terceiro quartil, pelo que:
Limite Inferior = Primeiro Quartil – 1,5 * (Terceiro Quartil – Primeiro Quartil)
Limite Superior = Terceiro Quartil + 1,5 * (Terceiro Quartil – Primeiro Quartil)

Assim, o objetivo principal do boxplot consiste na análise da distribuição dos dados. A partir dessa análise, é possível obter conclusões sobre o centro dos dados (representado pela mediana), a amplitude dos dados (calculada através da diferença entre o máximo e o mínimo), a simetria ou assimetria do conjunto de dados e a presença de outliers.

# Explicação teórica: Proporção
A proporção é uma medida que expressa a relação ou comparação entre dois conjuntos de dados. De modo geral, é expressa como uma fração, onde o numerador representa a parte ou subconjunto de interesse e o denominador representa o todo ou o conjunto total, ou seja, será obtido um valor compreendido entre 0 e 1.

É utilizada para descrever a distribuição entre os diferentes tipos de dados.
Em suma, permite entender a relação entre diferentes grupos de dados, expressando a parte de interesse em relação ao todo.

# Análise dos resultados obtidos através da proporção
A proporção indica a relação entre o número de indivíduos de um determinado grupo que recomendariam o parque a outras pessoas (respostas "Sim") e o número total de indivíduos desse grupo.

Proporção de crianças que recomendam o parque: 0.4732824427480916

Proporção de adultos que recomendam o parque: 0.5439560439560439

Proporção de séniores que recomendam o parque: 0.45454545454545453

Assim, é possível concluir que a faixa etária dos adultos tem a maior proporção de recomendação: com uma proporção de 54,40%, os adultos são os mais propensos a recomendar o parque comparativamente às outras faixas etárias.
Já as crianças e os séniores, têm uma taxa de aprovação idêntica, com 47% e 45% respetivamente.
Em suma,  os adultos têm uma proporção de recomendação cerca de 10% superior quando comparados com a faixa etária das crianças e dos séniores.
