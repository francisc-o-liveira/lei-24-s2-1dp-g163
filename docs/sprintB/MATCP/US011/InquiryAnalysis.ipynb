```python
import fileinput

Type1Yes = 0
Type2Yes = 0
Type3Yes = 0

Type1No = 0
Type2No = 0
Type3No = 0

Type1Times = 0
Type2Times = 0
Type3Times = 0

def switch_caseYes(argument, argument2):
    global Type1Yes, Type2Yes, Type3Yes, Type1Times, Type2Times, Type3Times
    if argument == "Child":
        Type1Yes += 1
        Type1Times += argument2
    elif argument == "Adult":
        Type2Yes += 1
        Type2Times += argument2
    elif argument == "Senior":
        Type3Yes += 1
        Type3Times += argument2
    else:
        print("Invalid case")

def switch_caseNo(argument, argument2):
    global Type1No, Type2No, Type3No, Type1Times, Type2Times, Type3Times
    if argument == "Child":
        Type1No += 1
        Type1Times += argument2
    elif argument == "Adult":
        Type2No += 1
        Type2Times += argument2
    elif argument == "Senior":
        Type3No += 1
        Type3Times += argument2
    else:
        print("Invalid case")

# User input for file name
filename = input("Enter the data file name (.csv): ")
scan = open(filename)

# Reading data
for line in scan:
    arrayLine = line.split()
    if arrayLine[1] == 'Y':
        switch_caseYes(arrayLine[0], int(arrayLine[2]))
    else:
        switch_caseNo(arrayLine[0], int(arrayLine[2]))

Type1= Type1Yes + Type1No
Type2= Type2Yes + Type2No
Type3= Type3Yes + Type3No

#Proporçao de Individuos que recomendam o parque por cada grupo etario
if(Type1 != 0):

    pType1Yes = (Type1Yes/Type1)
    print(f"The proportion of Childs who would recommend the park to others {pType1Yes}")

if(Type2 != 0):

    pType2Yes = (Type2Yes/Type2)
    print(f"The proportion of Adult who would recommend the park to others {pType2Yes}")

if(Type3 != 0):

    pType3Yes = (Type3Yes/Type3)
    print(f"The proportion of Senior who would recommend the park to others {pType3Yes}")

```
